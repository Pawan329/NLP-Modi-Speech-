## Function that converts list to srting

```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += (" "+ele)   
    
    # return string   
    return str1  
        
        
# Driver code     
s = speech_clean2
speech_clean3 = listToString(s)
```
